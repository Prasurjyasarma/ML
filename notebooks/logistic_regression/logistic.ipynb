```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import kagglehub
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```

```python
path = kagglehub.dataset_download("dileep070/heart-disease-prediction-using-logistic-regression")

print("Path to dataset files:", path)
```

```python
csv_file = "/root/.cache/kagglehub/datasets/dileep070/heart-disease-prediction-using-logistic-regression/versions/1/framingham.csv"
data = pd.read_csv(csv_file)
```

```python
data.isnull().sum()
data = data.fillna(data.mean())
```

```python
X = data.drop("TenYearCHD", axis = 1)
Y = data["TenYearCHD"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
model = LogisticRegression(max_iter=2000)
model.fit(x_train, y_train)

print(f"Model Intercept: {model.intercept_}")
print(f"Model slope: {model.coef_[0]}")
```

```python
y_pred = model.predict(x_test)
```

```python
print("Accuracy score :", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
