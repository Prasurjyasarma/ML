```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.cluster import KMeans
import kagglehub
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score
```

```python
path = kagglehub.dataset_download("shwetabh123/mall-customers")
print(path)
```

```python
data = pd.read_csv("/root/.cache/kagglehub/datasets/shwetabh123/mall-customers/versions/1/Mall_Customers.csv")
data.head(5)
```

```python
X = data[['Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
model = KMeans(n_clusters=5, random_state=42)
model.fit(X)

labels = model.labels_
data['cluster'] = labels
```

```python

plt.scatter(X['Annual Income (k$)'], X['Spending Score (1-100)'], c=labels)
plt.xlabel("Income")
plt.ylabel("Spending Score")
plt.show()
```

```python
new_point = [[10, 30]] 
predicted_cluster = model.predict(new_point)
print(predicted_cluster)
```

```python
score = silhouette_score(X, model.labels_)
print(score)
```
